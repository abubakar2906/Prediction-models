```python
import pandas as pd

cars_file_path = "cardekho_data.csv"
cars_data = pd.read_csv(cars_file_path)

# filtered_car_data = cars_data.dropna(axis=0)
```

```python
cars_data.head()
cars_data = cars_data.drop(columns=['Car_Name'],)
```

```python
cars_data['Fuel_Type'] = cars_data['Fuel_Type'].map({'Petrol': 0, 'Diesel': 1, 'CNG': 2})
cars_data['Seller_Type'] = cars_data['Seller_Type'].map({'Dealer': 0, 'Individual': 1})
cars_data['Transmission'] = cars_data['Transmission'].map({'Manual': 0, 'Automatic': 1})
cars_data.head()
```

```python
y = cars_data['Selling_Price']

cars_features = ['Year', 'Present_Price', 'Kms_Driven', 'Fuel_Type', 'Seller_Type', 'Transmission', 'Owner']
X = cars_data[cars_features]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_valid, y_train, y_valid = train_test_split(X, y, train_size=0.8, test_size=0.2, random_state=0)
```

```python
from sklearn.tree import DecisionTreeRegressor
```

```python
cars_model = DecisionTreeRegressor()
cars_model.fit(X_train, y_train)
```

```python
cars_model.predict(X_valid)
cars_model.score(X_valid, y_valid)
```
